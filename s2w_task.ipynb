# Loading the CSV file

```python
#loading the CSV file with help of CSVLoader from langchain
loader = CSVLoader(file_path='/content/south_indian_food_questions (1) (1).csv')
data=loader.load()
```

# Cleaning the and creating dictionary and list of question and answer

```python
# cleaning and creating dictionary and as question as key and answer as value and also create a list of questions to store in vectorstore

result_dict = {}
questions = []

for i in data:
    content = i.page_content.split('\n')
    question = content[0].replace('Question: ', '').strip()
    answer = content[1].replace('Answer: ', '').strip()

    result_dict[question] = answer
    questions.append(question)
```

```python
questions
```

# Embedding

```python
# use "sentence-transformers/all-MiniLM-L6-v2" model from HuggingFace to create embeddings of questions
embedding = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
```

# Storing the list of question in vectorstore

```python
# Using FAISS as vectorstore because it is faster for searching and free
db=FAISS.from_texts(questions,embedding)
```

# Function to perform similarity search and retrieve the answer

```python
#function for perform similarity search  with the query and the questions which stored in vectorstore.
#and retrieve one similar question from vectorstore and return the corresponding answer from a results dictionary.
def find_answer(query):
  similarity=db.similarity_search(query,k=1)
  similar_one=similarity[0].page_content
  return result_dict[similar_one],similar_one
```

```python
user_query = "what breakfast dish is most people like in south india which is made with rice cakes"

retrieved_answer,quest = find_answer(user_query)
retrieved_answer
```

# Enhanced Answer

```python
from getpass import getpass

# Prompt for the API key
google_api_key = getpass('Enter your Google API key: ')
```

```python
#using gemini model , Tried llama 3.2 3b model but its not performing as gemini
llm = GoogleGenerativeAI(model="gemini-pro", google_api_key=google_api_key)
```

```python
from langchain.prompts import PromptTemplate,ChatPromptTemplate

# providing prompt and generate an answer with gemini model
template = """generate and provide more information about the answer based on the quesion,
  if the answer is not the answer of the question don't try to make up an answer just respond as 'Unable to find the answer for that question from uploaded DOCUMENT',

  answer: {answer}

  question: {question}

  Helpful Answer:"""
custom_rag_prompt = PromptTemplate.from_template(template)

rag_chain = (
custom_rag_prompt
| llm
)

output=rag_chain.invoke({'answer':retrieved_answer,'question':quest})
```
